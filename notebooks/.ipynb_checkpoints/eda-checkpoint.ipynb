## Calculating the Hurst Exponent for India VIX and NIFTY50

```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
def hurst_exponent(ts):
    """
    Compute Hurst exponent using R/S analysis.
    H < 0.5 → mean-reverting
    H ≈ 0.5 → random walk
    H > 0.5 → trending
    """
    lags = range(2, 20)
    tau = []
    for lag in lags:
        # std of lagged differences
        tau.append(np.std(np.subtract(ts[lag:], ts[:-lag])))
        
    # Fit log-log regression: slope = H
    poly = np.polyfit(np.log(lags), np.log(tau), 1)
    return poly[0]


def rolling_hurst(series, window=120):
    return series.rolling(window).apply(lambda x: hurst_exponent(x.values), raw=False)

vix = yf.Ticker("^INDIAVIX").history(period="2y")["Close"].dropna()
vix.index = vix.index.tz_localize(None)

# Rolling Hurst
window = 120
vix_hurst = rolling_hurst(vix, window=window)
vix_rank = vix.rolling(252).apply(lambda x: pd.Series(x).rank(pct=True).iloc[-1], raw=False)
latest_vix    = vix.iloc[-1]
latest_hurst  = vix_hurst.iloc[-1]
latest_rank   = vix_rank.iloc[-1]

print(f"India VIX        : {latest_vix:.2f}")
print(f"VIX Percentile   : {latest_rank:.2%}")
print(f"Hurst Exponent   : {latest_hurst:.3f}")

if latest_rank > 0.5 and latest_hurst < 0.5:
    print("Premium is rich AND vol is mean-reverting: Sell")
elif latest_rank > 0.5 and latest_hurst >= 0.5:
    print("Premium is rich BUT vol is trending: trade smaller / wait")
elif latest_rank <= 0.5:
    print("Premium is thin: avoid or skip")
```

## Historical RV vs current IV

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------------------------
# 1) Download historical data
nifty_ticker = "^NSEI"
vix_ticker   = "^INDIAVIX"

nifty = yf.download(nifty_ticker, start="2015-01-01")
vix   = yf.download(vix_ticker, start="2015-01-01")

# -------------------------------------------------------------------
# 2) Keep only 'Close' and rename columns
nifty = nifty[['Close']].rename(columns={'Close':'NIFTY_Close'})
vix   = vix[['Close']].rename(columns={'Close':'VIX'})

# -------------------------------------------------------------------
# 3) Merge on date (inner join keeps only common dates)
df = nifty.merge(vix, left_index=True, right_index=True, how='inner')

# -------------------------------------------------------------------
# 4) Compute daily log returns for NIFTY
df['ret'] = np.log(df['NIFTY_Close'] / df['NIFTY_Close'].shift(1))

# -------------------------------------------------------------------
# 5) Compute 30-day forward realized volatility
window = 30
df['RV_30_fwd'] = df['ret'].rolling(window).std() * np.sqrt(252)
df['RV_30_fwd'] = df['RV_30_fwd'].shift(-window)  # forward-looking

# -------------------------------------------------------------------
# 6) Convert VIX to decimal
df['VIX'] = df['VIX'] / 100.0

# -------------------------------------------------------------------
# 7) Drop rows with missing forward RV
df = df.dropna(subset=['RV_30_fwd', 'VIX'])

# -------------------------------------------------------------------
# 8) Plot time series
plt.figure(figsize=(12,6))
plt.plot(df.index, df['VIX'], label='India VIX (Implied 30d Vol)')
plt.plot(df.index, df['RV_30_fwd'], label='30d Forward Realized Vol')
plt.legend()
plt.title('India VIX vs Forward 30-Day Realized Volatility (NIFTY50)')
plt.ylabel('Volatility (Annualized)')
plt.show()

# -------------------------------------------------------------------
# 9) Scatter plot
plt.figure(figsize=(8,8))
plt.scatter(df['VIX'], df['RV_30_fwd'], alpha=0.3)
plt.plot([0,0.5],[0,0.5], color='red', linestyle='--')  # 45° line
plt.xlabel('India VIX (Implied 30d Vol)')
plt.ylabel('30-Day Forward Realized Vol')
plt.title('Implied vs Realized Vol: India VIX vs NIFTY50')
plt.show()

# -------------------------------------------------------------------
# 10) Print stats
print("Average India VIX:", df['VIX'].mean())
print("Average 30d forward RV:", df['RV_30_fwd'].mean())
print("Average Volatility Risk Premium (VIX − RV):", (df['VIX'] - df['RV_30_fwd']).mean())
```
